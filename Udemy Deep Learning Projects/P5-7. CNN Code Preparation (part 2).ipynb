# CNN Arithmetic

**CNN Arithmetic**

$
H_{out}=
⌊{(H_{in} + 2 × P - D × (K - 1) - 1)/S + 1}⌋
$

D: Dilation

P: Padding

K: Kernel size

S: Stride

`H_out = np.floor((H_in + 2 * padding - dilation*(kernel_size - 1)-1)/stride +1)`

```python
import numpy as np
kernel_size = 2       # The size of the sliding window, must be > 0.
stride = kernel_size  # Default kernel_size - The stride of the sliding window, must be > 0. Default value is kernel_size
padding = 0           # Default 0 - Implicit negative infinity padding to be added on both sides, must be >= 0 and <= kernel_size / 2.
dilation = 1          # Default 1 - The stride between elements within a sliding window, must be > 0.

H_in = 100

H_out = np.floor((H_in + 2 * padding - dilation*(kernel_size - 1)-1)/stride +1)
print(H_out)
```
